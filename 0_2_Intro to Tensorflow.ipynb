## Understanding Tensor via Numpy

Tensor (`multi-dimensional array`) is an object that is invariant under a change of coordinate and has components that change in a special, predictable way under a change of coordinates.
\
A tensor does not depend on a coordinate system while its components can be changed if the system is replaced. (Refer to forward and backward transformations.)
\
We can considers tensors as partial derivatives and gradients that transform with the Jacobian Matrix.

## Variable

a special tensor which is useful to store states subject to change at any time..
\
\
We generate parameters such as weights and biases which need to be obtained by training via 'variable'.

```python
# Values in a Tensor are not supposed to be changed in the first place.
tensor = tf.ones((3, 4))
print(tensor)

tensor[0, 0] = 2.
```

```python
# in order to change the components of a tensor..
variable = tf.Variable(tensor)
print(variable)

variable[0, 0].assign(2)
print(variable)
```

```python
# possible to change the components of a tensor as follows..
initial_value = tf.random.normal(shape = (2, 2)) # np,random.randn
weight = tf.Variable(initial_value)
print(weight)
```

```python
# possible to change the components of a tensor using 'initializer'
weight = tf.Variable(tf.random_normal_initializer(stddev = 1.)(shape = (2, 2)))
print(weight)
```

```python
# renewing values of a variable via methods such as
# `.assign(value)`, `.assign_add(increment)`, or
# `.assign_sub(decrement)`

new_value = tf.random.normal(shape = (2, 2))
print(new_value)
weight.assign(new_value)
print(weight)
```

```python
added_value = tf.ones(shape = (2, 2))
weight.assign_sub(added_value)
print(weight)
```

Modelling 1 - Using Keras Sequential API (the easiest)

```python
# like stacking bricks
def create_seq_model():
    model = keras.Sequential()
    # single hidden layer and the output layer
    model.add(keras.layers.Flatten(input_shape = (28, 28)))
    model.add(keras.layers.Dense(128, activation = 'relu'))
    model.add(keras.layers.Dropout(0.2))
    model.add(keras.layers.Dense(10, activation = 'softmax'))
    return model
```

```python
seq_model = create_seq_model()
```

```python
# batch size not decided yet (None)
# 28 X 28  = 784
seq_model.summary()
```

Modelling 2 - Using Keras Functional API (more flexible approach than the above)

```python
# like a chain
def create_func_model():
    inputs = keras.Input(shape = (28, 28))
    flatten = keras.layers.Flatten()(inputs)
    dense = keras.layers.Dense(128, activation = 'relu')(flatten)
    drop = keras.layers.Dropout(0.2)(dense)
    outputs = keras.layers.Dense(10, activation = 'softmax')(drop)
    model = keras.Model(inputs = inputs, outputs = outputs)
    return model
```

```python
func_model = create_func_model()
```

```python
func_model.summary()
```

Modelling 3 - Using Model Class Subclassing (the most complex approach)  similar to PyTorch

```python
class SubClassModel(keras.Model):
    def __init__(self):
        super(SubClassModel, self).__init__()
        self.flatten = keras.layers.Flatten(input_shape = (28, 28))
        self.dense1 = keras.layers.Dense(128, activation = 'relu')
        self.drop = keras.layers.Dropout(0.2)
        self.dense2 = keras.layers.Dense(10, activation = 'softmax')
    
    # like a chain
    def call(self, x, training = False): # 'training': for dropout process
        x = self.flatten(x)
        x = self.dense1(x)
        x = self.drop(x)
        return self.dense2(x)
```

```python
subclass_model = SubClassModel()
```

```python
# Since summary method in this approach requires inserting inputs,
# before applying summary method, we need inputs.
inputs = tf.zeros((1, 28, 28))
subclass_model(inputs)
subclass_model.summary()
```

Training / Validation 2 - Using Gradient Tape

```python
# loss function
loss_object = keras.losses.CategoricalCrossentropy()

# optimizer
learning_rate = 0.001
optimizer = keras.optimizers.Adam(learning_rate = learning_rate)

# loss, accuracy
train_loss = keras.metrics.Mean(name = 'train_loss')
train_accuracy = keras.metrics.CategoricalAccuracy(name = 'train_accuracy')
test_loss = keras.metrics.Mean(name = 'test_loss')
test_accuracy = keras.metrics.CategoricalAccuracy(name = 'test_accuracy')
```

```python
# @: decorator
# After TensorFlow 2.x, due to introducing 'Eager' mode,
# the speed has dropped..
# Decorator can help TensorFlow to alleviate this problem
# by converting into graph mode (freezing).
# Keep in mind that 'print()' command does not work if under @.
# If you need to debug referring to printed contents,
# then at first write codes w/o @ and after all errors are removed add @.
# FYI: Keras API automatically uses the graphic mode.
@tf.function
def train_step(model, images, labels):
    with tf.GradientTape() as tape: # for backpropagation
        # training = True -> only needed if there are layers with
        # different behavior during training vs. inference (e.g. Dropout)
        predictions = model(images, training = True)
        loss = loss_object(labels, predictions)
    # differentiate loss w.r.t. every parameter (trainable variables)
    # gradient tape: recording from the input layer to getting loss value
    # and doing backpropagation during rewinding and referring to
    # the recorded parameters
    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    
    train_loss(loss)
    train_accuracy(labels, predictions)
```

```python
# test step
@tf.function
def test_step(model, images, labels):
    # training = False -> only needed if there are layers with
    # different behavior during training vs. inference (e.g. Dropout)
    predictions = model(images, training = False)
    t_loss = loss_object(labels, predictions)
    
    test_loss(t_loss)
    test_accuracy(labels, predictions)
```

```python
EPOCHS = 10

for epoch in range(EPOCHS):
    # Reset the metrics at the beginning of the next epoch.
    train_loss.reset_states()
    train_accuracy.reset_states()
    test_loss.reset_states()
    test_accuracy.reset_states()
    
    for images, labels in train_ds:
        train_step(func_model, images, labels)
    
    for test_images, test_labels in test_ds:
        train_step(func_model, test_images, test_labels)
        
    print(
        f'Epoch {epoch + 1}, '
        f'Loss: {train_loss.result()}, '
        f'Accuracy: {train_accuracy.result() *  100}, '
        f'Test Loss: {test_loss.result()}, '
        f'Test Accuracy: {test_accuracy.result() * 100}'
    )
```

Tensorboard for Visualization 2 - Using Summary Writer provided by TensorFlow

```python
%load_ext tensorboard
```

```python
new_model_2 = create_seq_model()

# loss fuction
loss_object = keras.losses.CategoricalCrossentropy()

# optimizer
learning_rate = 0.001
optimizer = keras.optimizers.Adam(learning_rate = learning_rate)

# loss, accuracy
train_loss = keras.metrics.Mean(name = 'train_loss')
train_accuracy = keras.metrics.CategoricalAccuracy(name = 'train_accuracy')

test_loss = keras.metrics.Mean(name = 'test_loss')
test_accuracy = keras.metrics.CategoricalAccuracy(name = 'test_accuracy')
```

```python
@tf.function
def train_step(model, images, labels):
    with tf.GradientTape() as tape:
        predictions = model(images, training = True)
        loss = loss_object(labels, predictions)
    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    
    train_loss(loss)
    train_accuracy(labels, predictions)
```

```python
@tf.function
def test_step(model, images, labels):
    predictions = model(images, training = False)
    t_loss = loss_object(labels, predictions)
    
    test_loss(t_loss)
    test_accuracy(labels, predictions)
```

```python
import datetime

current_time = datetime.datetime.now().strftime('%Y%m%d-%H%M%S')
train_log_dir = 'logs/gradient_tape/' + current_time + '/train'
test_log_dir = 'logs/gradient_tape' + current_time + '/test'
train_summary_writer = tf.summary.create_file_writer(train_log_dir)
test_summary_writer = tf.summary.create_file_writer(test_log_dir)
```

```python
EPOCHS = 10

for epoch in range(EPOCHS):
    train_loss.reset_states()
    train_accuracy.reset_states()
    test_loss.reset_states()
    test_accuracy.reset_states()
    
    for images, labels in train_ds:
        train_step(new_model_2, images, labels)
    with train_summary_writer.as_default():
        tf.summary.scalar('loss', train_loss.result(), step = epoch)
        tf.summary.scalar('accuracy', train_accuracy.result(), step = epoch)
        
    for test_images, test_labels in test_ds:
        test_step(new_model_2, test_images, test_labels)
    with test_summary_writer.as_default():
        tf.summary.scalar('loss', test_loss.result(), step = epoch)
        tf.summary.scalar('accuracy', test_accuracy.result(), step = epoch)
    
    print(
        f'Epoch {epoch + 1}, '
        f'Loss: {train_loss.result()}, '
        f'Accuracy: {train_accuracy.result() * 100}, '
        f'Test Loss: {test_loss.result()}, '
        f'Test Accuracy: {test_accuracy.result() * 100}'
        )
```
